```python
import dataUtils
import LongShortTermMemoryNetworks
import datetime as dt
```

```python
def initilizeModels(modelGroupName: str):
    
    LSTM1 = LongShortTermMemoryNetworks.LSTMmodel(lookBack=32,
                                                 architecture=[16],
                                                 dropout=0.0,
                                                 regularization=0.1,
                                                 bidirectional=False,
                                                 convInputLayer=False,
                                                 modelName=modelGroupName + "-Model-1")

    LSTM2 = LongShortTermMemoryNetworks.LSTMmodel(lookBack=32,
                                                 architecture=[16,16],
                                                 dropout=0.0,
                                                 regularization=0.1,
                                                 bidirectional=True,
                                                 convInputLayer=False,
                                                 modelName=modelGroupName + "-Model-2")

    LSTM3 = LongShortTermMemoryNetworks.LSTMmodel(lookBack=32,
                                                 architecture=[8],
                                                 dropout=0.0,
                                                 regularization=0.0001,
                                                 bidirectional=False,
                                                 convInputLayer=True,
                                                 kernelSize=4,
                                                 filters=8,
                                                 modelName=modelGroupName + "-Model-3",)
    
    return [LSTM1, LSTM2, LSTM3]
```

```python
splitDate = dt.datetime(2006, 1, 3)
endDate = dt.datetime(2008, 1, 3)
```

```python
assetList = ["AAPL"]
data = dataUtils.loadScaleDataMultivariate(assetList=assetList,
                                          loadpath="./Data/",
                                          endDate=endDate)

data.splitDataByDate(splitDate=splitDate)
data.createLSTMDataSet(lookBack=32)
for model in initilizeModels("1-Asset"):
    model.createModel(noTimeSeries=1, gpuOptmized=False)
    model.train(10000, data, learningRate=0.1,verbose=0,saveModel=True)
```

```python
assetList = ["WMT", "AAPL", "ABT"]
data = dataUtils.loadScaleDataMultivariate(assetList=assetList,
                                          loadpath="./Data/",
                                          endDate=endDate)

data.splitDataByDate(splitDate=splitDate)
data.createLSTMDataSet(lookBack=32)
for model in initilizeModels("3-Assets"):
    model.createModel(noTimeSeries=6, gpuOptmized=False)
    model.train(10000, data, learningRate=0.1,verbose=0,saveModel=True)
```
